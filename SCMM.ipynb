```python
import librosa
import os
import pandas as pd

def extract_features(file_path):
    try:
        audio, sample_rate = librosa.load(file_path)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        mfccs_processed = np.mean(mfccs.T, axis=0)
    except Exception as e:
        print("Error encountered while parsing file: ", file_path)
        return None
    return mfccs_processed

# Directory containing audio files
audio_dir = '/content/drive/MyDrive/MIREX/Audio'

# Dictionary to hold features and file names
features = []

# Iterate through each file in the folder
for file_name in os.listdir(audio_dir):
    if file_name.endswith('.mp3'):
        file_path = os.path.join(audio_dir, file_name)
        data = extract_features(file_path)
        if data is not None:
            features.append([file_name.replace('.mp3', ''), data])


```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ContextRepresentationModule(nn.Module):
    def __init__(self, lyric_dim, audio_dim, num_heads):
        super(ContextRepresentationModule, self).__init__()
        self.lyric_global_attention = nn.MultiheadAttention(lyric_dim, num_heads)
        self.audio_global_attention = nn.MultiheadAttention(audio_dim, num_heads)
        self.lyric_gru = nn.GRU(lyric_dim, lyric_dim // 2, bidirectional=True, batch_first=True)
        self.audio_gru = nn.GRU(audio_dim, audio_dim // 2, bidirectional=True, batch_first=True)
        self.lyric_direct_mapping = nn.Linear(lyric_dim, lyric_dim)
        self.audio_direct_mapping = nn.Linear(audio_dim, audio_dim)

    def forward(self, lyrics, audio):

        print(f"Input lyrics shape (before transpose): {lyrics.shape}")
        print(f"Input audio shape (before transpose): {audio.shape}")
        # Transpose for MultiheadAttention
        lyrics_transposed = lyrics.transpose(0, 1)
        audio_transposed = audio.transpose(0, 1)

        print(f"Input lyrics shape (after transpose): {lyrics_transposed.shape}")
        print(f"Input audio shape (after transpose): {audio_transposed.shape}")

        # Apply MultiheadAttention
        global_lyric_context, _ = self.lyric_global_attention(lyrics_transposed, lyrics_transposed, lyrics_transposed)
        global_audio_context, _ = self.audio_global_attention(audio_transposed, audio_transposed, audio_transposed)

        # Transpose back to batch first
        global_lyric_context = global_lyric_context.transpose(0, 1)
        global_audio_context = global_audio_context.transpose(0, 1)

        # Apply GRU layers
        local_lyric_context, _ = self.lyric_gru(global_lyric_context)
        local_audio_context, _ = self.audio_gru(global_audio_context)

        direct_lyric_map = self.lyric_direct_mapping(global_lyric_context)
        direct_audio_map = self.audio_direct_mapping(global_audio_context)

        return global_lyric_context, local_lyric_context, direct_lyric_map, global_audio_context, local_audio_context, direct_audio_map











class FullInteraction(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(FullInteraction, self).__init__()
        self.fc = nn.Linear(input_dim, output_dim)

    def forward(self, audio, lyrics):
        concatenated = torch.cat((audio, lyrics), dim=-1)
        return self.fc(concatenated)


class SelfAdaptivePathSelection(nn.Module):
    def __init__(self, feature_dim):
        super(SelfAdaptivePathSelection, self).__init__()
        self.query = nn.Parameter(torch.randn(1, feature_dim))
        self.fc = nn.Linear(feature_dim, feature_dim)

    def forward(self, full_interaction):
        # Compute scores as cosine similarity
        scores = F.cosine_similarity(full_interaction, self.query, dim=1)
        # Convert scores to weights
        weights = F.softmax(scores, dim=0).unsqueeze(1)
        # Apply weights to paths
        weighted = full_interaction * weights
        return self.fc(weighted.sum(dim=0))

# Define the ModalInteractionModule without partial and biased interactions
class ModalInteractionModule(nn.Module):
    def __init__(self, audio_dim, lyrics_dim, output_dim):
        super(ModalInteractionModule, self).__init__()
        self.full_interaction = FullInteraction(audio_dim + lyrics_dim, output_dim)
        self.path_selection = SelfAdaptivePathSelection(output_dim)

    def forward(self, audio, lyrics):
        # Perform the full interaction
        full_interaction = self.full_interaction(audio, lyrics)
        # Perform path selection
        selected_features = self.path_selection(full_interaction)
        return

class CombinedModel(nn.Module):
    def __init__(self, lyric_dim, audio_dim, num_heads, output_dim, num_categories, num_clusters):
        super(CombinedModel, self).__init__()
        self.context_module = ContextRepresentationModule(lyric_dim, audio_dim, num_heads)
        self.modal_interaction_module = ModalInteractionModule(audio_dim, lyric_dim, output_dim)
        # Additional layers to generate category and cluster predictions
        self.category_output = nn.Linear(output_dim, num_categories)
        self.cluster_output = nn.Linear(output_dim, num_clusters)

    def forward(self, lyrics, audio):
        # Get context representations
        global_lyric_context, local_lyric_context, direct_lyric_map, global_audio_context, local_audio_context, direct_audio_map = self.context_module(lyrics, audio)

        # Combine the outputs of the context module
        combined_context = torch.cat([global_lyric_context, local_lyric_context, direct_lyric_map, global_audio_context, local_audio_context, direct_audio_map], dim=-1)

        # Pass through the modal interaction module
        multimodal_features = self.modal_interaction_module(combined_context)

        # Generate category and cluster predictions
        category_pred = self.category_output(multimodal_features)
        cluster_pred = self.cluster_output(multimodal_features)

        return category_pred, cluster_pred

# Assuming dimensions and number of heads
lyric_dim = 100  # The dimension of the lyric embeddings
audio_dim = 40   # The dimension of the audio features
num_heads = 5    # Number of heads for multi-head attention
output_dim = 128 # Output dimension for full interaction

# Initialize the context representation module
context_module = ContextRepresentationModule(lyric_dim, audio_dim, num_heads)

# Initialize the modal interaction module
modal_interaction_module = ModalInteractionModule(audio_dim, lyric_dim, output_dim)

# Convert input tensors to float32
padded_audio_features = padded_audio_features.float()
padded_lyric_embeddings = padded_lyric_embeddings.float()

# Forward pass through the module
multimodal_features = modal_interaction_module(padded_audio_features, padded_lyric_embeddings)

multimodal_features
```

```python
import torch
from torch.utils.data import DataLoader, Dataset
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import ast

def chunk_sequence(sequence, chunk_size):
    return [sequence[i:i + chunk_size] for i in range(0, len(sequence), chunk_size)]

chunk_size = 50

def string_to_list(string):
    try:
        return ast.literal_eval(string)
    except:
        raise ValueError("Could not convert string to list")

class LyricsAudioDataset(Dataset):
    def __init__(self, lyrics_embeddings, audio_features, categories, clusters, chunk_size):
        self.chunk_size = chunk_size

        # Convert string representations to lists if necessary
        self.lyrics_embeddings = [string_to_list(le) for le in lyrics_embeddings]
        self.audio_features = [string_to_list(af) for af in audio_features]

        # Chunking the sequences and expanding categories and clusters to match
        self.lyrics_embeddings_chunks = []
        self.audio_features_chunks = []
        self.categories_labels = []
        self.clusters_labels = []

        for idx, _ in enumerate(self.lyrics_embeddings):
            lyrics_chunks = chunk_sequence(self.lyrics_embeddings[idx], self.chunk_size)
            audio_chunks = chunk_sequence(self.audio_features[idx], self.chunk_size)

            # Assume there's only one category and cluster label per sequence
            category_label = categories[idx]
            cluster_label = clusters[idx]

            # Extend the dataset lists
            self.lyrics_embeddings_chunks.extend(lyrics_chunks)
            self.audio_features_chunks.extend(audio_chunks)
            self.categories_labels.extend([category_label] * len(lyrics_chunks))
            self.clusters_labels.extend([cluster_label] * len(audio_chunks))

    def __len__(self):
        return len(self.lyrics_embeddings_chunks)

    def __getitem__(self, idx):
        lyrics_tensor = torch.tensor(self.lyrics_embeddings_chunks[idx], dtype=torch.float32)
        audio_tensor = torch.tensor(self.audio_features_chunks[idx], dtype=torch.float32)
        category_tensor = torch.tensor(self.categories_labels[idx], dtype=torch.long)
        cluster_tensor = torch.tensor(self.clusters_labels[idx], dtype=torch.long)

        return lyrics_tensor, audio_tensor, category_tensor, cluster_tensor

# Usage
train_dataset = LyricsAudioDataset(
    train_data['Lyric_embeddings'].tolist(),
    train_data['Audio_Features'].tolist(),
    train_data['Category'].tolist(),
    train_data['Cluster'].tolist(),
    chunk_size
)
val_dataset = LyricsAudioDataset(
    val_data['Lyric_embeddings'].tolist(),
    val_data['Audio_Features'].tolist(),
    val_data['Category'].tolist(),
    val_data['Cluster'].tolist(),
    chunk_size
)


print(f"Padded lyric embeddings shape: {padded_lyric_embeddings.shape}")
print(f"Padded audio features shape: {padded_audio_features.shape}")

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)


# Initialize the model and other components for training
num_epochs = 100
lyric_dim = 100
audio_dim = 40
num_heads = 10 # adjusted number of heads
output_dim = 128
num_categories = len(set(dataset_info['Category']))
num_clusters = len(set(dataset_info['Cluster']))

model = CombinedModel(lyric_dim, audio_dim, num_heads, output_dim, num_categories, num_clusters)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training and Validation Loop
for epoch in range(num_epochs):
    model.train()
    for lyrics, audio, category_labels, cluster_labels in train_loader:
      print(lyrics.shape)
      print(audio.shape)
      optimizer.zero_grad()
      category_pred, cluster_pred = model(lyrics, audio)
      loss = criterion(category_pred, category_labels) + criterion(cluster_pred, cluster_labels)
      loss.backward()
      optimizer.step()

    # Validation phase
    model.eval()
    all_category_preds, all_cluster_preds = [], []
    all_category_labels, all_cluster_labels = [], []
    with torch.no_grad():
        for lyrics, audio, category_labels, cluster_labels in val_loader:
          category_pred, cluster_pred = model(lyrics, audio)
          all_category_preds.extend(category_pred.argmax(dim=1).tolist())
          all_cluster_preds.extend(cluster_pred.argmax(dim=1).tolist())
          all_category_labels.extend(category_labels.tolist())
          all_cluster_labels.extend(cluster_labels.tolist())

    # Compute metrics
    category_accuracy = accuracy_score(all_category_labels, all_category_preds)
    cluster_accuracy = accuracy_score(all_cluster_labels, all_cluster_preds)
    category_precision = precision_score(all_category_labels, all_category_preds, average='macro', zero_division=0)
    cluster_precision = precision_score(all_cluster_labels, all_cluster_preds, average='macro', zero_division=0)
    category_recall = recall_score(all_category_labels, all_category_preds, average='macro')
    cluster_recall = recall_score(all_cluster_labels, all_cluster_preds, average='macro')
    category_f1 = f1_score(all_category_labels, all_category_preds, average='macro')
    cluster_f1 = f1_score(all_cluster_labels, all_cluster_preds, average='macro')

    print(f'Epoch {epoch+1}/{num_epochs}')
    print(f'Category - Accuracy: {category_accuracy}, Precision: {category_precision}, Recall: {category_recall}, F1 Score: {category_f1}')
    print(f'Cluster - Accuracy: {cluster_accuracy}, Precision: {cluster_precision}, Recall: {cluster_recall}, F1 Score: {cluster_f1}')
```
